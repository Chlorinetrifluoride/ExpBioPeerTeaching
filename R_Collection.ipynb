## Plotting with basic R

```python

# Define/load your date, e.g in a dataframe, here dataframe with the columns A and B containing 1000 random entries
data <- data.frame(A = rnorm(1000, mean = 0, sd = 1),B = rnorm(1000, mean = 1, sd = 1))
```

### General

#### Labelling

```python

# To label a R-base plot pass the arguments main for the title and xlab/ylab for the axis labels
hist(data$A, main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
```

#### Adding some colour

```python

## To add some colour to your life, choose your favorite and pass it along
## RGB values with transparency can be passed in the form: col=rgb(0, 0, 0, 0.15)
hist(data$A, col = "salmon",main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
```

#### Adding a legend


```python

hist(data$A, col = "salmon",main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
# so everyone knows what you are doing, create a legend, passing position,
# a string/vector/list... for the labels and
# a vector with colornames/rgb objects
legend("topright", legend = "A", fill = c("salmon"))
```

#### Specify limits

```python

# set (non)sensible limits based on your data by passing xlim and ylim arguments
# containing vectors with upper and lower bounds:
hist(data$A,xlim=c(-1,1),ylim=c(0,50),col = "salmon",main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
```

#### Arranging multiple plots

```python

# To plot multiple plots use the par function:
par(mfrow = c(1, 2))  # 1 row, 2 columns
hist(data$A,col = "salmon",main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
hist(data$B,col = "lightgreen",main = "Histogram of more Random \n Data", xlab = "Values", ylab = "Counts")
```

```python

# You can also use the same figure for multiple plots:
## To facilitate interpretation of overlayed plots it could be useful to vary
## the density
## Set labels and title in the first call to the plotting Function:
hist(data$A, col = "salmon",density = 80,main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
## For additional calls set the argument add =TRUE
hist(data$B, col = "lightgreen",density = 50, add = TRUE)
```

### Histogram

```python

# A basic Histogram can be plotted by passing a numeric vector to hist,
# setting freq=FALSE plots the probability density
par(mfrow = c(1, 2))  # 1 row, 2 columns
hist(data$A,freq=TRUE,col ="salmon",main = "Histogram of Random Data", xlab = "Values", ylab = "Counts")
hist(data$A,freq=FALSE,col ="salmon",main = "Histogram of Random Data", xlab = "Values", ylab = "Density")
```

#### Binning

```python

# To add custom bins pass a vector containing the desired bins as the breaks argument,
# the easiest way to do this is the built-in seq function, where you
# pass the lower bound of your date, the upper bound and the bin width
# take care that the bounds are higher/lower than the highest/lowest values
hist(data$A,breaks = seq(-4, 4, by = 0.5), main = "Histogram of Random Data", xlab = "Values", ylab = "Frequency",col = "salmon")
```

At the center of the histogram is the size and following the number of bins. Choosing the right size for your bins is important and there is no ultimate recipe.
Factors to consider:
* distribution of your data
* what you you want to know

Choosing bins with an higher width reduces noise, but could hide features, while a lower width shows more details, but brings in noise.

## Plotting with ggplot

```python

library(ggplot2)
```

ggplot() Function:

You start building a ggplot by using the ggplot() function. This function initializes the plot and specifies the dataset to use.

```python
ggplot(data = your_data)
```

Aesthetics (aes):

Aesthetics map variables in your dataset to visual elements in the plot (e.g., x-axis, y-axis, color, shape). The aes() function is used to define these mappings.

## Basic Histogram
1. a) Create a basic Histogram with the values of the normal distribution on    the x-axis andf the frequency of them on the y-axis. \\
b) Define a colour as well a fill for the bars. \\
c) Define a binwidth of 0.5. \\
d) Label the graph as well as the x- and y-axis


```python


data1 <- data.frame(values = rnorm(100, mean = 0, sd = 1))

# Define in aes that x should be the actual values from the dataframe and ggplot knows that y should be the frequency of the values
ggplot(data1, aes(x = values)) +
  # Define the binwidth, fill, color, alpha (transperancy)
  geom_histogram(binwidth = 0.5, fill = "skyblue", color = "black", alpha = 0.7) +
  # Labels für Grah, X- und Y-Achse
  labs(title = "Histograms with ggplot2", x = "Values", y = "Frequency")
```

## Arranging Plots

### Now we want to visualize multiple datasets in one plot

2. a) Find a way to put both data1 and data2 into one ggplot. \\
b) Define a colour as well a fill for the bars. \\
c) Define a binwidth of 0.5. \\
d) Label the graph as well as the x- and y-axis \\
e) Change the y-axis to display density instead of frequency (you don't have to calculate the density values)



```python

data1 <- data.frame(value = rnorm(1000, mean = 0, sd = 1))
data2 <- data.frame(value = rnorm(1000, mean = 2, sd = 2))

ggplot() +
  geom_histogram(data = data1, aes(x = value),
                 binwidth = 0.5, fill = "skyblue", colour = "black", alpha = 0.7, position = "identity") +
  geom_histogram(data = data2, aes(x = value),
                 binwidth = 0.5, fill = "coral", colour = "white", alpha = 0.7, position = "identity") +
  labs(title = "Overlaid Histograms", x = "Value", y = "Count")
```
